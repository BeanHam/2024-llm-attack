```python
import os
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Easy Case

```python
data_path = 'inference_results'
epoch_index = {
    'checkpoint-126':0.1,
    'checkpoint-252':0.2,
    'checkpoint-378':0.3,
    'checkpoint-504':0.4,
    'checkpoint-630':0.5,
    'checkpoint-756':0.6,
    'checkpoint-882':0.7,
    'checkpoint-1008':0.8,
    'checkpoint-1134':0.9,
    'checkpoint-1260':1.0,
}
files = os.listdir(f'../{data_path}/')
files.remove('.ipynb_checkpoints')
files = np.unique([f.split('.')[0] for f in files])
files = files[files!='final_model']
files
```

```python
ACCURACY = []
HAMMING = []
EPOCH = []
for file in files:
    epoch = epoch_index[file]
    with open(f'../{data_path}/{file}.json') as f: d = json.load(f)    
    EPOCH.append(epoch)    
    ACCURACY.append(d['accuracy'])
    HAMMING.append(d['hamming'])
vis_data = pd.DataFrame(np.c_[EPOCH, ACCURACY, HAMMING], columns=['epoch', 'accuracy', 'hamming'])
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))
ax2 = ax1.twinx()
sns.lineplot(data=vis_data, x="epoch", y="accuracy", ax=ax1, color='dodgerblue', label='accuracy')
sns.lineplot(data=vis_data, x="epoch", y="hamming", ax=ax2, color='darkorange', label='confidence')
ax1.set_ylabel('Accuracy')
ax2.set_ylabel('Hamming Distance')
plt.legend()
plt.show()
pass
```
